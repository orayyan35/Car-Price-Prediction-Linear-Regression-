```python
duplicates = df['car_ID'].duplicated()
duplicates.sum()
```

```python
carname = df['CarName'].str.split(' ').str[0]
df.insert(3,"carname",carname)
df.drop(['CarName'],axis=1,inplace=True)
df.head(n=10)
```

```python
df.loc[df['carname'] == 'nissan', 'carname'] = 'Nissan'
df.loc[df['carname'] == 'maxda', 'carname'] = 'mazda'
df.loc[df['carname'] == 'porcshce', 'carname'] = 'porsche'
df.loc[df['carname'] == 'vokswagen', 'carname'] = 'Volkswagen'
df.loc[df['carname'] == 'vw', 'carname'] = 'Volkswagen'
df.loc[df['carname'] == 'volkswagen', 'carname'] = 'Volkswagen'
df.loc[df['carname'] == 'toyouta', 'carname'] = 'toyota'
df.carname.unique()
```

```python
plt.figure(figsize=(20,8))
plt.plot(df['price'])
plt.title('Car Price')
plt.show()
```
